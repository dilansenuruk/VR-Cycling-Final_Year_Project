```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#resistive values for time instances for the path
res_dic = {'1': 0.5 , '20': 5 , '52': 0.5 , '54': -4, '73': 1.4  , '84': -1.3 , '90': -5, '93': 3.3,
           '104': -0.4, '109':-3.1, '113': 3.3, '137':-3.2,'142': 0.95, '145': -1.3, '149': -2.2,
           '154':-3.1, '163':-4, '172': 3.4, '174': 0.5, '179':-4, '191':2.4, '196':0.5, '220':-3.2,
           '227':1.4, '230':-0.4, '240':0.5, '252':-1.05, '258':0.95, '261':-1.05, '270':0.5,
           '293':5, '317':0.5, '326':-3.55}
```

P = F x V

Resistive_power (F) = P / V

```python
# k values are adding to the F so that the user will feel a resistance
time_values = [i for i in range(0, 340)]
k_val = 0
k_val_list = []

for j in time_values:
    if (str(j) in res_dic.keys()):
        k_val = res_dic[str(j)]
    k_val_list.append(k_val)

fig, ax = plt.subplots(1, 1, figsize=(15, 3))
ax.plot(time_values, k_val_list)
ax.set_xlabel("time in video (in seconds)")
ax.set_ylabel("k value")
ax.grid("on")
plt.show()
```

```python
# approximated path for the assigned k value sequence
integrated_list = []
value = 0
for j in k_val_list:
    value += j
    integrated_list.append(value)

fig, ax = plt.subplots(1, 1, figsize=(15, 3))
ax.plot(time_values, integrated_list)
ax.set_xlabel("time in video (in seconds)")
ax.set_ylabel("integrated k value")
ax.grid("on")
plt.show()
```

```python
# approximated path for the assigned k value sequence (with flat paths)
integrated_list = []
value = 0
for j in k_val_list:
    value += (j - 0.5)
    integrated_list.append(value)

fig, ax = plt.subplots(1, 1, figsize=(15, 3))
ax.plot(time_values, integrated_list)
ax.set_xlabel("time in video (in seconds)")
ax.set_ylabel("integrated k value")
ax.grid("on")
plt.show()
```

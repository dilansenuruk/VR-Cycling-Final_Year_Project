```python
import pandas as pd
import csv
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('BLE_Delay.csv')
print(df.to_string())
```

```python
delay_array=[]
for id, row in df.iterrows():
    delay_array.append(row["BLE delays"]/2)
print("Mean Delay =",np.mean(delay_array),"us")
print("Max Delay =", max(delay_array),"us")
print("Min Delay =", min(delay_array),"us")
print("Median Delay =", np.median(delay_array),"us")
print("std = ",np.std(delay_array))
#print(np.mean(delay_array))

#delay_array = [x for x in delay_array if x<=150]
plt.figure(figsize=(10,6))
plt.xlabel('Delay in microsecond')
plt.ylabel('Frequency')
plt.title('BLE Histogram')
plt.hist(delay_array, bins=50, color='skyblue', edgecolor='black')
plt.show()
```
